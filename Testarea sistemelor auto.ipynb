```python
import time
import random

def testeaza_sisteme_auto(durata_planificata_secunde):
    sisteme = ["Motor", "Sistem Frânare", "Direcție", "Sistem Electric", "Transmisie"]
    start_time = time.time()
    timp_scurs = 0
    defect_detectat = False

    print(f"--- Începe testarea sistemelor (Durată planificată: {durata_planificata_secunde}s) ---")

    # Bucla rulează cât timp nu am depășit timpul și nu avem defecte
    while timp_scurs < durata_planificata_secunde and not defect_detectat:
        timp_actual = time.time()
        timp_scurs = timp_actual - start_time
        
        # Alegem un sistem la întâmplare pentru a-l monitoriza în acest pas
        sistem_verificat = random.choice(sisteme)
        
        # Simulăm o probabilitate de defect (ex: 5% șanse să apară o eroare)
        stare_sistem = random.choices(["OK", "DEFECT"], weights=[95, 5])[0]
        
        if stare_sistem == "DEFECT":
            print(f"[ALARMĂ] Defect detectat la: {sistem_verificat} după {timp_scurs:.2f} secunde!")
            defect_detectat = True
        else:
            print(f"[INFO] {timp_scurs:.2f}s: Verificare {sistem_verificat}... Status: {stare_sistem}")
        
        # Mică pauză între verificări pentru a nu supraîncărca procesorul
        time.sleep(0.5)

    # Concluzia testului
    print("\n--- Rezumat Test ---")
    if defect_detectat:
        print("STATUS: TEST EȘUAT. Vehiculul necesită mentenanță.")
    else:
        print(f"STATUS: TEST FINALIZAT CU SUCCES. Toate sistemele sunt funcționale după {durata_planificata_secunde}s.")

# Rulăm testul pentru 10 secunde
testeaza_sisteme_auto(10)
```
